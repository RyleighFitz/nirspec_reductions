```python
def peak2d(array):
    '''This function finds the peak in a 2D array by the
       recursive method.

       Complexity: O(n log m)'''

    n = len(array)
    m = len(array[0])

    j = m//2

    row = [i[j] for i in array]

    i = row.index(max(row))

    #print(i, j)

    if j > 0 and array[i][j] < array[i][j-1]:
        return peak2d([row[:j] for row in array]), i, j

    elif j < m - 1 and array[i][j] < array[i][j+1]:
        return peak2d([row[j:] for row in array]), i, j

    else:
        return array[i][j], i, j
    
```

```python
# TO DO: get accurate r_pix
#for now we'll estimate r_pix = 25. NEED to make this more accurate!

def titan_ortho_coords(img, titan_cen, r_pix=25., alpha=0):
    """ Take a scam image of Titan and convert it to orthographic coordinates.
    img: scam Image
    radius: Titan radius (that night) in arcsec
    """
    
    #spat_scale = hr_scams[0].header['SPATSCAL']
    #r_pix = spat_scale*radius
    
    #get image coordinates [-1:1, disk centered at (a,b) = (0,0)]
    cx,cy = titan_cen[0]
    i,j = np.meshgrid(range(img.header['NAXIS1']), range(img.header['NAXIS2']))
    a = (j-cy)/25.     
    b = (cx-i)/25.

    #Rotate based on slit angle in scam imaged
    if alpha != 0:
        alpha = np.radians(90-alpha)
        rot_matrix = np.array(( (np.cos(alpha), -np.sin(alpha)),
                        (np.sin(alpha),  np.cos(alpha)) ))

        rotA = np.einsum('ji, mni -> jmn', rot_matrix, np.dstack([a,b]))
        b_ = rotA[0]
        a_ = rotA[1]
        
    return a_,b_
```

```python
def slit_ortho_coords(img, titan_cen, alpha, r_pix=25.):
    """ Return orthonormal coordinates of top and bottom of slit (on Titan's disk)."""
    #Find intersection (edge of titan and slit)
    m = np.tan(np.radians(90-alpha))
    slity, slitx = img.header['SLITX'], img.header['SLITY']
    b =slity - m*slitx

    cx,cy = titan_cen[0]

    d = np.sqrt(((r_pix**2)*(1+m**2) - (cx-m*cy-b)**2))
    x1, y1 = (cy+cx*m-b*m + d)/(1+m**2), (b+cy*m+cx*(m**2) +m*d)/(1+m**2)
    x2, y2 = (cy+cx*m-b*m - d)/(1+m**2), (b+cy*m+cx*(m**2) -m*d)/(1+m**2)
    
    rot_matrix = np.array(( (np.cos(np.radians(90-alpha)), -np.sin(np.radians(90-alpha))),
                        (np.sin(np.radians(90-alpha)),  np.cos(np.radians(90-alpha))) ))

    rot1 = np.einsum('ji, mni -> jmn', rot_matrix, np.dstack([(x1-cy)/25.,(cx-y1)/25.]))
    nx1 = rot1[0]
    ny1= rot1[1]

    rot2 = np.einsum('ji, mni -> jmn', rot_matrix, np.dstack([(x2-cy)/25.,(cx-y2)/25.]))
    nx2 = rot2[0]
    ny2= rot2[1]

    return (nx1,ny1), (nx2,ny2)
```

```python
def get_corners_from_cen(ortho_x, ortho_y):
    x_delta = 1/(0.8/0.041) #Based on slit width
    new_xs = np.array([[x - x_delta, x + x_delta] for x in ortho_x])

    y_delta = (ortho_y[1] - ortho_y[0])/2. #Calculated from y centers
    ortho_y = ortho_y - 2*y_delta
    new_ys = np.array([[y + y_delta, y+y_delta] for y in ortho_y])#.reshape((17,2))
    
    return new_xs, new_ys
```
